# Cek dan Penanganan Outlier

```python
# Boxplot
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik yang ingin dicek outliernya
num_cols = ['Age', 'SibSp', 'Parch', 'Fare']

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

d. Apakah outlier pada data tersebut perlu diatasi serta bagaimana cara mengatasinya?

Jawab :

Penanganan Outlier perlu diatasi berdasarkan jumlah dan dampak outlier, dengan menggunakan metode Winsorizing (Mengganti Outlier dengan Batas IQR) karena jika data penting dan tidak boleh dihapus, kita bisa mengganti outlier dengan batas kuartil.

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

**a.** Jelaskan proses pembagian data menjadi data training dan testing dan alasan mengapa perlu membagi data tersebut
Jawab :     
**Proses Pembagian Dataset**:    
1.   Pengumpulan Data yaitu dengan mengumpulkan dataset yang relevan untuk analisis.
2.   Pembersihan Data yaitu dengan membersihkan data dari nilai yang hilang dan kesalahan.
3.   Pemisahan data menjadi data X (fitur) dan data y (outcome)
4.    Pembagian dataset yaitu dengan menggunakan metode *Train-Test Split* misalnya, 80% untuk training dan 20% untuk testing
    *  Training Set digunakan untuk melatih model. Model belajar dari data atau fitur-fitur yang relevan tanpa 'terganggu' oleh label saat pelatihan ini untuk menemukan pola dan hubungan.
    *  Testing Set digunakan untuk menguji model setelah dilatih. Ini membantu menilai seberapa baik model dapat memprediksi data yang belum pernah dilihat sebelumnya.

**Alasan mengapa perlu membagi data**
1.  Untuk menghindari Overfitting karena jika model dilatih dan diuji pada data yang sama, terdapat risiko model akan terlalu cocok dengan data tersebut (overfitting), maka tidak dapat berfungsi dengan baik pada data baru.
2.   Mengevaluasi performa model secara objektif, karena outcome digunakan sebagai acuan kebenaran untuk mengukur akurasi dan efektivitas prediksi model.





**b.** Berapa persen pembagiannya? Mengapa memilih membagi dengan proporsi tersebut?

Jawab :      
Digunakan proporsi 80:20.
Artinya: 80% data digunakan untuk training (melatih model), 20% data digunakan untuk testing (mengukur kinerja model).

Alasan Memilih Proporsi 80:20:
*   Menghindari Overfitting
Dengan menggunakan 80% data untuk melatih, model bisa belajar dari pola data yang beragam. Jika data training terlalu sedikit, model akan overfit (menghafal data, bukan memahami pola).
*     Evaluasi Akurat
Testing set sebesar 20% sudah cukup representatif untuk menilai kemampuan model terhadap data baru yang belum pernah dilihat.

**d.** Berapa nilai evaluasi dari masing-masing model pada data uji? Apa arti dari masing-masing metrik evaluasi tersebut?

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

df = pd.read_csv('Titanic-Dataset.csv')

features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
target = 'Survived'

for col in ['Sex', 'Embarked']:
    df[col] = LabelEncoder().fit_transform(df[col])

imputer = SimpleImputer(strategy='mean')
df[features] = imputer.fit_transform(df[features])

scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])

X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Inisialisasi model
nb_model = GaussianNB()
knn_model = KNeighborsClassifier(n_neighbors=5)
nn_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)

# Latih dan evaluasi model
models = {'Naive Bayes': nb_model, 'KNN': knn_model, 'Neural Network': nn_model}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"=== Evaluasi {name} ===")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", recall_score(y_test, y_pred, average='weighted'))
    print("F1-Score:", f1_score(y_test, y_pred, average='weighted'))
    print(classification_report(y_test, y_pred))
    print("\n")
```
